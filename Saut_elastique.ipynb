```python
# Importation de la librairie ncessaire pour les tracs
%matplotlib
from matplotlib.pyplot import* 
from matplotlib.widgets import Cursor

# Dfinition des constantes g et m

# Constante physique : g = 9.81 m/s
g=9.81

# Paramtre : masse du sauteur en kg
m=100

# Cration des listes vides

h=[]      # liste pour stocker les valeurs des hauteurs de chute
Epp=[]    # liste pour stocker les valeurs des nergies potentielles de pesanteur

# Ajout itratif des valeurs de hauteurs de chute dans la liste h

for i in range(0,31,1):   # les valeurs s'tendent de 0 m  31 m exclu par pas de 1 m 
    h.append(i)            # ajout d'un lment  la fin de la liste h

# Calcul et ajout itratif des valeurs de l'nergie potentielle de pesanteur Epp(J) correspondantes aux hauteurs de chute
 
for i in range(0,31,1):    # les valeurs s'tendent de 0 m  31 m exclu par pas de 1 m 
    Epp.append(-m*g*h[i])  # ajout d'un lment  la fin de la liste Epp
    
 
# Trac de l'nergie potentielle de pesanteur et de l'nergie cintique en fonction de la hauteur
plot(h,Epp,label="Epp")      # plot(abscisse,ordonne,"legende de la courbe")
xlabel("Hauteur de chute (m)")       # xlabel("Nom de l'axe de l'axe des absicces")
ylabel("Energies (J)")               # ylabel("Nom de l'axe des ordonnes")
legend(loc="lower left")             # Position de la legende de la courbe dans le graphique
cursor = Cursor(gca(), color="red", lw=2) # Affiche un curseur sur le graphique
grid(True)                           # Affiche la grille du graphique
show(True)                           # Affiche le graphique
```
